```python
ind=df_can.loc["India",years]
ind
```

```python
# immigration from different countries

df_CI=df_can.loc[['India','China'], years]
df_CI
```

```python
df_CI.head()
```

```python
df_can.sort_values(by="Total",ascending=False,axis=0, inplace=True)
df_can.head()
```
